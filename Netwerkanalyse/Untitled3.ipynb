```python
import pandas as pd

MOTnetwerk=pd.read_csv('MOTNetwerk.csv', sep=';',error_bad_lines=False,header=None ,encoding='utf-8')
```

```python
MOTnetwerkListofLists= MOTnetwerk.values.tolist()
```

# P1
Schrijf een python programma dat  uit dit bestand van moties een <strong>ongericht</strong> NetworkX netwerk maakt, waarbij de knopen gelabeld zijn met hun partij.

> Twee personen A en B zijn verbonden precies dan als er een motie bestaat waar A indiener is en B medeindiener, en eentje waar juist B indiener is en A medeindiener.

```python
indieners = {i[0]: i[1] for i in MOTnetwerkListofLists[1:]}
medeindieners = {i[2]: i[3] for i in MOTnetwerkListofLists[1:]}

P = nx.Graph()

for key, value in indieners.items():    
    P.add_node(key, partij=value)

for key, value in medeindieners.items():    
    P.add_node(key, partij=value)
    
paren = [(i[0], i[2]) for i in MOTnetwerkListofLists[1:]]

for a, b in paren:
    P.add_edge(a, b)
    
nx.draw(P)
```
