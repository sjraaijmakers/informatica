# P1
Schrijf een python programma dat  uit dit bestand van moties een <strong>ongericht</strong> NetworkX netwerk maakt, waarbij de knopen gelabeld zijn met hun partij.

> Twee personen A en B zijn verbonden precies dan als er een motie bestaat waar A indiener is en B medeindiener, en eentje waar juist B indiener is en A medeindiener. 

         
         

```python
indieners = {i[0]: i[1] for i in MOTnetwerkListofLists[1:]}
medeindieners = {i[2]: i[3] for i in MOTnetwerkListofLists[1:]}

P = nx.Graph()

all = indieners.copy()
all.update(medeindieners)

for key, value in all.items():    
    P.add_node(key, partij=value)
    
paren = [(i[0], i[2]) for i in MOTnetwerkListofLists[1:]]

for a, b in paren:
    P.add_edge(a, b)
```

De datums in deze file zijn (bijna) in het zogeheten [iso-date formaat](https://en.wikipedia.org/wiki/ISO_8601). Zoek het op en snap de ideeen erachter.

4. Beschouw je lijst met datums als knopen in een netwerk. De relatie tussen knopen is de _eerder dan_ relatie. Wat is de [dichtheid (density)](https://en.wikipedia.org/wiki/Network_science#Density)?

```python
data = [i[4] for i in MOTnetwerkListofLists[1:]]

print sorted(data)
```

Het valt op dat de sort() functie zomaar toegepast kan worden op data. Er moet is dus ergens in de python broncode gedefinieerd zijn wat het sorteren van twee (of meerdere) data inhoudt.

Waarschijnlijk snapt de python interpreter dat bij het vergelijken van twee data eerst gekeken moet worden naar het jaartal. Wanneer A.jaar < B.jaar of A.jaar > B.jaar kunnen we concluderen dat A < B of A > B. Wanneer de jaren echter gelijk zijn moet er gekeken worden naar de maand, wanneer de maand wéér hetzelfde is kan er gekeken worden naar de dag om uiteindelijk de conclusie te kunnen trekken of A < B of A > B (of A == B).

```python
data = sorted(list(set(data)))

D = nx.Graph()

for d in data:
    D.add_node(d)    

for i in data:
    for j in data:
        if i != j:
            if i < j:
                D.add_edge(i, j)
                
print nx.density(D)
```
