# Q19
Maak een nieuw ongericht netwerk waarin je sterke en zwakke edges onderscheid, door te zeggen dat een edge tussen A en B sterk is als zowel (A,B) als (B,A) bestond (dus A heeft B een keer gevraagd, en ook andersom). Als er alleen maar een pijl 1 kant op is noemen we die edge zwak. Bepaal welke knopen voldoen aan de Strong Triadic Closure property. Wat zegt dit?

```python
A = nx.Graph()

# add nodes
for old in H.nodes():
    A.add_node(old)

# add edges
for a, b in H.edges():    
    if H.has_edge(b, a):
        A.add_edge(a, b, weight=0.9)
    else:
        A.add_edge(a, b, weight=0.1)
        
# Definitie van "Sterke link"
def is_strong(weight):
    if weight > 0.5:
        return True
    return False

# STC Checking
def STC_h(G, x):
    # Vraag naar alle neighbors van X
    neighbors = G.neighbors(x)
    strongs = []
    for n in neighbors:
        weight = G.get_edge_data(x, n)["weight"]
        if is_strong(weight):
            # Voeg neighbor toe aan strong list
            strongs.append(n)

    # Alle unieke combinaties van de elementen in "strong" met een lengte van 2
    combinations = list(itertools.combinations(strongs, 2))

    for i, j in combinations:
        # Check of unieke combinatie uberhaupt een linnk heeft
        if not G.has_edge(i, j):
            # Zo niet; return dan dit paar (als tegenvoorbeeld)
            return (i, j)
    # Als alle combinaties dus wel een edge hebben geld de STC:
    return True

stcs = []
for node in A.nodes():
    if STC_h(A, node):
        stcs.append(A)
```
